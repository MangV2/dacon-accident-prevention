```python
from datasets import Dataset

# combined_training_dataê°€ Pandas DataFrameì¸ ê²½ìš°
dataset_hf = Dataset.from_pandas(combined_training_data.reset_index(drop=True))
```

```python
prompt_template = """
### ì§€ì¹¨: ë‹¹ì‹ ì€ ê±´ì„¤ ì•ˆì „ ì „ë¬¸ê°€ì…ë‹ˆë‹¤.
ì§ˆë¬¸ì— ëŒ€í•œ **ì¬ë°œ ë°©ì§€ ëŒ€ì±… ë° í–¥í›„ ì¡°ì¹˜ ê³„íš**ë§Œ ê°„ê²°í•˜ê²Œ ë‹µë³€í•˜ì„¸ìš”.
- ì„œë¡ , ë°°ê²½ ì„¤ëª…, ì¶”ê°€ ì„¤ëª… ì—†ì´ í•µì‹¬ ë‚´ìš©ë§Œ ì „ë‹¬í•˜ì„¸ìš”.
- ë¶ˆí•„ìš”í•œ í˜•ì‹(ëª©ì°¨, ê°•ì¡° í‘œì‹œ, ë¦¬ìŠ¤íŠ¸ ë“±)ì„ ì‚¬ìš©í•˜ì§€ ë§ˆì„¸ìš”.
- í•œ ë¬¸ì¥ ë˜ëŠ” ê°„ê²°í•œ ë¬¸ë‹¨ìœ¼ë¡œ ìì—°ìŠ¤ëŸ½ê²Œ ì‘ì„±í•˜ì„¸ìš”.
- íŠ¹ìˆ˜ë¬¸ìë¥¼ í¬í•¨í•˜ì§€ ë§ˆì„¸ìš”.

{context}

### ì§ˆë¬¸:
{question}

### ë‹µë³€:
"""
```

```python
def generate_prompt(data_point):
    # ì¶”ê°€ contextê°€ í•„ìš”í•œ ê²½ìš° ì—¬ê¸°ì„œ ì„¤ì • (í˜„ì¬ëŠ” ë¹ˆ ë¬¸ìì—´)
    context = ""
    question = data_point["question"]
    # prompt_templateì„ ì‚¬ìš©í•´ promptë¥¼ ìƒì„±í•©ë‹ˆë‹¤.
    prompt = prompt_template.format(context=context, question=question)
    # ëª¨ë¸ ì…ë ¥ìœ¼ë¡œ ì‚¬ìš©í•  í”„ë¡¬í”„íŠ¸ì™€ ì •ë‹µ(í‰ê°€ìš©)ë§Œ ë°˜í™˜í•©ë‹ˆë‹¤.
    return {"prompt": prompt, "answer": data_point["answer"]}

# ê¸°ì¡´ ë°ì´í„°ì…‹ì˜ ëª¨ë“  ì»¬ëŸ¼ì„ ì œê±°í•˜ê³ , generate_promptì—ì„œ í•„ìš”í•œ ì»¬ëŸ¼ë§Œ ë‚¨ê¹ë‹ˆë‹¤.
remove_column_keys = list(dataset_hf.features.keys())
dataset_cvted = dataset_hf.shuffle().map(generate_prompt, remove_columns=remove_column_keys)
```

```python
def tokenize_function(examples):
    outputs = tokenizer(examples["prompt"], truncation=True, max_length=512)
    return outputs
```

```python
remove_column_keys = dataset_cvted.features.keys()
dataset_tokenized = dataset_cvted.map(tokenize_function, batched=True, remove_columns=remove_column_keys)
```

```python
dataset_tokenized
```

```python
# í† í¬ë‚˜ì´ì§• ì™„ë£Œëœ ë°ì´í„°ì…‹ì„ train/testë¡œ ë¶„ë¦¬
split_datasets = dataset_tokenized.train_test_split(test_size=0.1, seed=42)

train_dataset = split_datasets["train"]
eval_dataset = split_datasets["test"]  # ë³´í†µ ê²€ì¦ ìš©ë„ë¡œ ì‚¬ìš©
```

```python
def collate_fn(examples):
    examples_batch = tokenizer.pad(examples, padding='longest', return_tensors='pt')
    examples_batch['labels'] = examples_batch['input_ids'] # ëª¨ë¸ í•™ìŠµ í‰ê°€ë¥¼ ìœ„í•œ loss ê³„ì‚°ì„ ìœ„í•´ ì…ë ¥ í† í°ì„ ë ˆì´ë¸”ë¡œ ì‚¬ìš©
    return examples_batch
```

```python
# 2. TrainingArgumentsì— evaluation_strategyì™€ eval_steps ì¶”ê°€
train_args = transformers.TrainingArguments(
    per_device_train_batch_size=8,         # ê° ë””ë°”ì´ìŠ¤ë‹¹ ë°°ì¹˜ ì‚¬ì´ì¦ˆ
    gradient_accumulation_steps=4,
    warmup_steps=1,
    max_steps=2000,
    learning_rate=2e-4,                     # í•™ìŠµë¥ 
    bf16=True,                              # bf16 ì‚¬ìš© (í•˜ë“œì›¨ì–´ ì§€ì› í™•ì¸ í•„ìš”)
    output_dir="outputs",
    optim="paged_adamw_8bit",               # 8ë¹„íŠ¸ AdamW ì˜µí‹°ë§ˆì´ì €
    logging_steps=50,                       # ë¡œê¹… ë¹ˆë„
    save_total_limit=3,                     # ì €ì¥í•  ì²´í¬í¬ì¸íŠ¸ì˜ ìµœëŒ€ ìˆ˜
    evaluation_strategy="steps",            # ì¼ì • ìŠ¤í…ë§ˆë‹¤ í‰ê°€ ì‹¤í–‰
    eval_steps=50                           # í‰ê°€ ì£¼ê¸° (50 ìŠ¤í…ë§ˆë‹¤ í‰ê°€)
)

# 3. SFTTrainerì— eval_datasetì„ ì¶”ê°€í•˜ì—¬ ìˆ˜ì •
trainer = SFTTrainer(
    model=model,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    args=train_args,
    data_collator=collate_fn
)
```

```python
model.config.use_cache = False
trainer.train()
```

```python
import pandas as pd
from datasets import Dataset
from tqdm.auto import tqdm

# 1. CSV íŒŒì¼ ë¡œë“œ (ì˜ˆ: test.csv)
test_df = pd.read_csv("/content/drive/MyDrive/test.csv")

# ë°ì´í„° ì „ì²˜ë¦¬
test_df['ê³µì‚¬ì¢…ë¥˜(ëŒ€ë¶„ë¥˜)'] = test_df['ê³µì‚¬ì¢…ë¥˜'].str.split(' / ').str[0]
test_df['ê³µì‚¬ì¢…ë¥˜(ì¤‘ë¶„ë¥˜)'] = test_df['ê³µì‚¬ì¢…ë¥˜'].str.split(' / ').str[1]
test_df['ê³µì¢…(ëŒ€ë¶„ë¥˜)'] = test_df['ê³µì¢…'].str.split(' > ').str[0]
test_df['ê³µì¢…(ì¤‘ë¶„ë¥˜)'] = test_df['ê³µì¢…'].str.split(' > ').str[1]
test_df['ì‚¬ê³ ê°ì²´(ëŒ€ë¶„ë¥˜)'] = test_df['ì‚¬ê³ ê°ì²´'].str.split(' > ').str[0]
test_df['ì‚¬ê³ ê°ì²´(ì¤‘ë¶„ë¥˜)'] = test_df['ì‚¬ê³ ê°ì²´'].str.split(' > ').str[1]

# 2. train ë°ì´í„°ì™€ ë™ì¼í•œ ë°©ì‹ìœ¼ë¡œ prompt ìƒì„±
def generate_prompt_from_row(row):
    prompt = (
        f"ê³µì‚¬ì¢…ë¥˜ ëŒ€ë¶„ë¥˜ '{row['ê³µì‚¬ì¢…ë¥˜(ëŒ€ë¶„ë¥˜)']}', ì¤‘ë¶„ë¥˜ '{row['ê³µì‚¬ì¢…ë¥˜(ì¤‘ë¶„ë¥˜)']}' ê³µì‚¬ ì¤‘ "
        f"ê³µì¢… ëŒ€ë¶„ë¥˜ '{row['ê³µì¢…(ëŒ€ë¶„ë¥˜)']}', ì¤‘ë¶„ë¥˜ '{row['ê³µì¢…(ì¤‘ë¶„ë¥˜)']}' ì‘ì—…ì—ì„œ "
        f"ì‚¬ê³ ê°ì²´ '{row['ì‚¬ê³ ê°ì²´(ëŒ€ë¶„ë¥˜)']}'(ì¤‘ë¶„ë¥˜: '{row['ì‚¬ê³ ê°ì²´(ì¤‘ë¶„ë¥˜)']}')ì™€ ê´€ë ¨ëœ ì‚¬ê³ ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤. "
        f"ì‘ì—… í”„ë¡œì„¸ìŠ¤ëŠ” '{row['ì‘ì—…í”„ë¡œì„¸ìŠ¤']}'ì´ë©°, ì‚¬ê³  ì›ì¸ì€ '{row['ì‚¬ê³ ì›ì¸']}'ì…ë‹ˆë‹¤. "
        f"ì¬ë°œ ë°©ì§€ ëŒ€ì±… ë° í–¥í›„ ì¡°ì¹˜ ê³„íšì€ ë¬´ì—‡ì¸ê°€ìš”?"
    )
    return {"prompt": prompt}

# DataFrameì˜ ê° í–‰ì— ëŒ€í•´ prompt ìƒì„± (ìƒˆë¡œìš´ ì—´ë¡œ ì¶”ê°€)
test_df = test_df.apply(generate_prompt_from_row, axis=1)
test_df = pd.DataFrame(test_df.tolist())

# 3. HuggingFace Datasetìœ¼ë¡œ ë³€í™˜
test_dataset = Dataset.from_pandas(test_df.reset_index(drop=True))

# (ì„ íƒì‚¬í•­) í† í¬ë‚˜ì´ì¦ˆ ì ìš© - ë³´í†µ inferenceì—ì„œëŠ” ì›ë³¸ promptë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.
# ë§Œì•½ tokenizerë¥¼ ì´ìš©í•œ ì „ì²˜ë¦¬ê°€ í•„ìš”í•˜ë‹¤ë©´ ì•„ë˜ì™€ ê°™ì´ map ì ìš©
remove_column_keys = test_dataset.features.keys()
test_dataset_tokenized = test_dataset.map(tokenize_function, batched=True, remove_columns=remove_column_keys)

# 4. ëª¨ë¸ íŒŒì´í”„ë¼ì¸ì„ ì‚¬ìš©í•´ ê²°ê³¼ ìƒì„±
# ë°°ì¹˜ ë‹¨ìœ„ë¡œ ì¶”ë¡ í•˜ë©´ì„œ tqdmìœ¼ë¡œ ì§„í–‰ ìƒí™©ì„ ëª¨ë‹ˆí„°ë§í•©ë‹ˆë‹¤.
batch_size = 16
prompts = test_dataset["prompt"]
results = []

for i in tqdm(range(0, len(prompts), batch_size), desc="Generating outputs"):
    batch_prompts = prompts[i:i+batch_size]
    batch_results = pipe(
        batch_prompts,
        do_sample=False,
        top_k=0,
        repetition_penalty=1.2,
        add_special_tokens=True
    )
    results.extend(batch_results)

# 5. ê²°ê³¼ ì¶œë ¥
for res in results:
    print(res)
```
